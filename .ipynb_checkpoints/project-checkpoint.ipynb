```python
ax = sns.countplot(x=df['machine_status'],order=df['machine_status'].value_counts(ascending=False).index);

abs_values = df['machine_status'].value_counts(ascending=False).values

ax.bar_label(container=ax.containers[0], labels=abs_values)
sns.set()
```

Letâ€™s plot machine status as function of time

```python
plt.plot(df['machine_status'])
```

From the above plot we can see that Recovering state follows Broken state.

  In Recovering state pump is trying to recover from Broken state



```python
##resample timestamp as a day
df['timestamp'] = pd.to_datetime(df['timestamp'])
df1 = df.set_index('timestamp') 
da = df1.resample(rule='D').mean()
```

mean of each sensor in each day from april to august

```python
df2 = df1.drop(['machine_status'], axis=1)
names=df2.columns
import warnings
for name in names:
    warnings.filterwarnings("ignore")
    sns.set()
    _ = plt.figure(figsize=(9,5))
    _ = da[name].plot()
    _ = plt.title(name)
```

plot the means of the readings for each state of the pump and see how different the means are from one condition to another i.e. normal vs broken state.

```python
normal = df[df['machine_status']=='NORMAL']
broken = df[df['machine_status']=='BROKEN']
recovering = df[df['machine_status']=='RECOVERING']
```

```python
## Extract mean values of the sensor readings from the pump in all the states
means_normal=normal.mean(axis=0)
means_broken = broken.mean(axis=0)
means_recovering = recovering.mean(axis=0)
```

```python
# Plot mean values of the sensor readings for each state of the pump

plt.figure(figsize=(18,9))
plt.plot(means_normal, marker='D', color='red', label='normal', linestyle='none')
plt.plot(means_broken, marker='D', color='blue', label='broken', linestyle='none')
plt.plot(means_recovering, marker='D', color='green', label='recovering', linestyle='none')
plt.xlabel('Sensors')
plt.ylabel('Mean Values')
plt.title('Means of the readings from each sensor')
plt.legend(loc='upper right')
plt.xticks(rotation=90)
plt.show()
```

some sensors have great difference in the means among each state

  sensor_04,sensor_28,sensor_31 have great differences
